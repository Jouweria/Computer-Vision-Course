```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#display the images 

def display(im_path):
        
        dpi = 80
        im_data = plt.imread(im_path)

        height, width  = im_data.shape[:2]
    
        # What size does the figure need to be in inches to fit the image?
        figsize = width / float(dpi), height / float(dpi)

        # Create a figure of the right size with one axes that takes up the full figure
        fig = plt.figure(figsize=figsize)
        ax = fig.add_axes([0, 0, 1, 1])

        # Hide spines, ticks, etc.
        ax.axis('off')

        # Display the image.
        ax.imshow(im_data, cmap='gray')

        plt.show()
```

# DRAWING WINDOW CIRCLE

```python
img = np.zeros((500,500,3))

def circle(event,x,y,radius,color):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,0,255),-1)
        
cv2.namedWindow(winname = 'Window')
cv2.setMouseCallback('Window',circle)
while True:
    cv2.imshow('Window',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```

# DRAWING A RECTANGLE CONSISTENTLY

```python
img2 = np.zeros((500,500,3))


ix = -1
iy = -1
iv,iz = -1,-1
drawing = False
drawing2 = False

def draw_rectangle(event,x,y,flags,params):
    global ix,iy,drawing,drawing2
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        drawing = True
        ix,iy = x,y
        
      
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(img2,(ix,iy),(x,y),(0,0,255),-1)
   
            
            
    elif event == cv2.EVENT_LBUTTONUP:
           
        drawing = False
        cv2.rectangle(img2,(ix,iy),(x,y),(0,0,255),-1)         
        
        
    elif event == cv2.EVENT_RBUTTONDOWN:
        
        drawing2 = True
        ix,iy = x,y    
            
           
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing2 == True:
            cv2.rectangle(img2,(ix,iy),(x,y),(0,255,0),-1)    
            
        
    elif event == cv2.EVENT_RBUTTONUP:
        
        drawing = False
        cv2.rectangle(img2,(ix,iy),(x,y),(0,255,0),-1)
       

    
cv2.namedWindow(winname = 'rectangle')
cv2.setMouseCallback('rectangle',draw_rectangle)

while True:
    cv2.imshow('rectangle',img2)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()   
```
